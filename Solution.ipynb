```python
import numpy as np
!unzip /content/anwer_sheet_samples.zip
from google.colab.patches import cv2_imshow
import cv2
img =cv2.imread("/content/anwer_sheet_samples/A001.png")
img1=cv2.imread("/content/anwer_sheet_samples/A002.png")
img2=cv2.imread("/content/anwer_sheet_samples/A003.png")
lower_threshold1 = np.uint8([130, 15 ,2])
upper_threshold1 = np.uint8([170, 255 ,255])
lower_threshold2 = np.uint8([170, 20 ,5])
upper_threshold2 = np.uint8([180, 255 ,250])
```

```python

img3=img
hsv = cv2.cvtColor(img3, cv2.COLOR_BGR2HSV)
mask1 = cv2.inRange(hsv, lower_threshold1, upper_threshold1)
mask2 = cv2.inRange(hsv, lower_threshold2, upper_threshold2)
markings = cv2.bitwise_and(img3,img3, mask= mask2)

hist=np.mean(mask1,axis=0)
indx=np.argmax(hist)
mask1[:,0:(indx-25)] = 0
mask1[:,(indx+25):]  = 0
margins = cv2.bitwise_and(img3,img3, mask= mask1)
cv2_imshow(margins)
```

```python
cv2_imshow(markings)
```
